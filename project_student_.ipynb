```python
import pandas as pd

# Load the dataset
data = pd.read_csv("students.csv")  # Adjust file path as needed

# Split the single column into multiple columns
data_cleaned = data.iloc[:, 0].str.split(";", expand=True)

# Assign column names
data_cleaned.columns = ["StudentID", "Name", "Age", "Email", "Department", "GPA", "GraduationYear"]

# Convert data types
data_cleaned["Age"] = pd.to_numeric(data_cleaned["Age"], errors='coerce')
data_cleaned["GPA"] = pd.to_numeric(data_cleaned["GPA"], errors='coerce')
data_cleaned["GraduationYear"] = pd.to_numeric(data_cleaned["GraduationYear"], errors='coerce')

# Inspect the cleaned data
df = pd.DataFrame(data_cleaned)
df
```

**What is the average GPA and age of students?**

```python
average_gpa = data_cleaned["GPA"].mean()
average_age = data_cleaned["Age"].mean()
print(f"Average GPA: {average_gpa:.2f}, Average Age: {average_age:.2f}")
```

**maximum and minimum GPA **




```python
# Find the maximum and minimum GPA
max_gpa = data_cleaned["GPA"].max()
min_gpa = data_cleaned["GPA"].min()

# Print the results
print(f"Maximum GPA: {max_gpa}")
print(f"Minimum GPA: {min_gpa}")
```

 **How many students are in the "Physics" department with a GPA greater than 3.5?**

```python
physics_high_gpa = data_cleaned[(data_cleaned["Department"] == "Physics") & (data_cleaned["GPA"] > 3.5)]
print(f"Number of Physics students with GPA > 3.5: {len(physics_high_gpa)}")
```

**How many students are younger than 20 and have a GPA above 3.0?**

```python
young_high_gpa_students = data_cleaned[(data_cleaned["Age"] < 20) & (data_cleaned["GPA"] > 3.0)]
print(f"Number of students younger than 20 with GPA > 3.0: {len(young_high_gpa_students)}")
```

```python
math_grads_2026 = data_cleaned[(data_cleaned["Department"] == "Mathematics") & (data_cleaned["GraduationYear"] == 2026)]
print(f"Number of Mathematics students graduating in 2026: {len(math_grads_2026)}")
```

```python
low_gpa_students = data_cleaned[data_cleaned["GPA"] < 2.5]
print(f"Number of students with GPA < 2.5: {len(low_gpa_students)}")
print("Details of these students:")
print(low_gpa_students[["StudentID", "Name", "GPA"]])
```

```python
physics_older_students = data_cleaned[(data_cleaned["Department"] == "Physics") & (data_cleaned["Age"] > 22)]
print(f"Number of Physics students older than 22: {len(physics_older_students)}")
```

```python
duplicate_gpa_diff_dept = data_cleaned.duplicated(subset=["GPA"], keep=False) & ~data_cleaned.duplicated(subset=["GPA", "Department"], keep=False)
students_duplicate_gpa = data_cleaned[duplicate_gpa_diff_dept]
print("Students with same GPA in different departments:")
print(students_duplicate_gpa)
```

```python
perfect_gpa_young = data_cleaned[(data_cleaned["Age"] == 18) & (data_cleaned["GPA"] == 4.0)]
print("Students aged 18 with GPA = 4.0:")
print(perfect_gpa_young[["StudentID", "Name", "GPA"]])
```

```python
average_gpa_by_department = data_cleaned.groupby("Department")["GPA"].mean()
print(average_gpa_by_department)
```

```python
students_per_department = data_cleaned.groupby("Department")["StudentID"].count()
print("Number of students in each department:")
print(students_per_department)
```

```python
max_gpa_per_department = data_cleaned.groupby("Department")["GPA"].max()
print("Highest GPA in each department:")
print(max_gpa_per_department)
```

```python
average_age_per_department = data_cleaned.groupby("Department")["Age"].mean()
print("Average age of students in each department:")
print(average_age_per_department)
```

```python
average_gpa_by_department_gradyear = data_cleaned.groupby(["Department", "GraduationYear"])["GPA"].mean()
print("Average GPA by department and graduation year:")
print(average_gpa_by_department_gradyear)
```

```python
total_gpa_per_department = data_cleaned.groupby("Department")["GPA"].sum()
print("Total GPA of students in each department:")
print(total_gpa_per_department)
```

```python
top_students = data_cleaned.sort_values("GPA", ascending=False).head(5)
print(top_students[["StudentID", "Name", "GPA"]])
```

```python
graduation_gpa_summary = data_cleaned[data_cleaned["GraduationYear"] >= 2025] \
    .groupby("Department").agg({"StudentID": "count", "GPA": "mean"})
graduation_gpa_summary.rename(columns={"StudentID": "StudentCount"}, inplace=True)
print(graduation_gpa_summary)
```

```python
import matplotlib.pyplot as plt

average_gpa_by_department.plot(kind="bar", figsize=(8, 5), color="skyblue")
plt.title("Average GPA by Department")
plt.ylabel("GPA")
plt.xlabel("Department")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(data_cleaned["GPA"], bins=10, kde=True, color="skyblue")
plt.title("Distribution of GPA")
plt.xlabel("GPA")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
data_cleaned["Department"].value_counts().plot(kind="bar", color="orange")
plt.title("Number of Students in Each Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_gpa_by_gradyear = data_cleaned.groupby("GraduationYear")["GPA"].mean()

plt.figure(figsize=(8, 5))
avg_gpa_by_gradyear.plot(kind="line", marker="o", color="green")
plt.title("Average GPA by Graduation Year")
plt.xlabel("Graduation Year")
plt.ylabel("Average GPA")
plt.grid(True)
plt.show()
```

```python
avg_gpa_by_department = data_cleaned.groupby("Department")["GPA"].mean()

plt.figure(figsize=(8, 5))
avg_gpa_by_department.plot(kind="bar", color="purple")
plt.title("Average GPA by Department")
plt.xlabel("Department")
plt.ylabel("Average GPA")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=data_cleaned["Age"], color="cyan")
plt.title("Age Distribution of Students")
plt.xlabel("Age")
plt.show()
```

```python
grad_year_counts = data_cleaned["GraduationYear"].value_counts()

plt.figure(figsize=(8, 5))
grad_year_counts.sort_index().plot(kind="bar", color="salmon")
plt.title("Number of Students by Graduation Year")
plt.xlabel("Graduation Year")
plt.ylabel("Count")
plt.show()
```

```python
high_gpa_counts = data_cleaned[data_cleaned["GPA"] > 3.5].groupby("Department")["StudentID"].count()

plt.figure(figsize=(8, 5))
high_gpa_counts.plot(kind="bar", color="gold")
plt.title("Students with GPA > 3.5 by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
dept_counts = data_cleaned["Department"].value_counts()

plt.figure(figsize=(8, 5))
dept_counts.plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Percentage Distribution of Students by Department")
plt.ylabel("")  # Remove y-axis label for better visualization
plt.show()
```

```python
avg_gpa_by_dept_gradyear = data_cleaned.groupby(["Department", "GraduationYear"])["GPA"].mean().unstack()

plt.figure(figsize=(10, 6))
avg_gpa_by_dept_gradyear.plot(kind="bar", stacked=True, colormap="viridis")
plt.title("Average GPA by Department and Graduation Year")
plt.xlabel("Department")
plt.ylabel("Average GPA")
plt.xticks(rotation=45)
plt.legend(title="Graduation Year")
plt.show()
```
